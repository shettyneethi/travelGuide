```python
import urllib3
from bs4 import BeautifulSoup
import json
import pymongo

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["DATASET"]
reviewsCol = mydb["REVIEWS"]
weatherCol = mydb["WEATHER"]
locations = reviewsCol.find({},{"LOCATION":1,"STATE":1})
locSet = set()

for l in locations:
    locSet.add(l['LOCATION']+"_"+l['STATE'])
print(len(locSet))
for location in locSet:
    location,state = location.split("_")
    tempDict = {"LOCATION":location,"STATE":state}
    place =location.replace(" ","%20")
    url = 'https://api.opencagedata.com/geocode/v1/json?q='+place+','+state+'&key=e9371220b0d64866ab25b1760944394f'
    print(url)
    http = urllib3.PoolManager()
    response = http.request('GET', url)
    soup = BeautifulSoup(response.data,"lxml")
    newDictionary=json.loads(soup.text)
#     print(newDictionary)
    lat = '%.3f'%(newDictionary["results"][0]['geometry']['lat'])
    lng = '%.3f'%(newDictionary["results"][0]['geometry']['lng'])
    q = str(lat)+","+str(lng)
    print(q)
    url = 'http://api.worldweatheronline.com/premium/v1/weather.ashx?key=db4faa45fbee429691655232192803&q='+q+'&format=json&mca=yes'
    response = http.request('GET', url)
    soup = BeautifulSoup(response.data,"lxml")
    newDictionary=json.loads(soup.text)
    tempDict["AVG_WEATHER"] = newDictionary['data']['ClimateAverages'][0]['month']
#     tempDict["AVG_WEATHER"] = newDictionary['data']
    
    print(tempDict["AVG_WEATHER"] )
    x = weatherCol.insert_one(tempDict)
    print("--------------------------------------------------")
```
